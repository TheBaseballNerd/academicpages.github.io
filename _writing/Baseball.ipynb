```python
import altair as alt
import pandas as pd
from pybaseball import playerid_lookup
from pybaseball import statcast_batter
```

```python
playerid_lookup('tatis','fernando')
```

```python
tatis_statcast_batter = statcast_batter('2019-03-28', '2019-09-29', 665487)
pd.set_option('display.max_columns', 85)
```

```python
tatis_statcast_batter.tail()
```

```python
tatis_statcast_batter.columns
```

```python
goatstats = tatis_statcast_batter[['launch_speed','launch_angle','estimated_woba_using_speedangle','woba_value','iso_value']]
```

```python
goatstats
```

```python
goatstatsnonull = goatstats.dropna()
```

```python
goatstatsnonull.head(50)
```

```python
goatfinalstats = goatstatsnonull.iloc[::-1]
```

```python
goatfinalstats.shape
```

```python
alt.Chart(goatfinalstats).mark_point(color='red').encode(
    x='launch_speed',
    y='launch_angle',
    color='woba_value'
)
```
